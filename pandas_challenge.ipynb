```python
#Gender Demographics
#Percentage and Count of Male Players
#Percentage and Count of Female Players
#Percentage and Count of Other/Non-Disclosed

df_male = df.loc[df["Gender"] == "Male", :] 
#print(df_male)

df_female = df.loc[df["Gender"] == "Female", :]
#print(df_female)

df_other = df.loc[df["Gender"] == "Other / Non-Disclosed", :]
#print(df_other)

#print(len(df_male["SN"].unique()))
#print(len(df_female["SN"].unique()))
#print(len(df_other["SN"].unique()))

gender_total = len(df_male["SN"].unique()) + len(df_female["SN"].unique()) + len(df_other["SN"].unique())

male_per = "{:.2%}".format(len(df_male["SN"].unique()) / gender_total)
female_per = "{:.2%}".format(len(df_female["SN"].unique()) / gender_total)
other_per = "{:.2%}".format(len(df_other["SN"].unique()) / gender_total)

Gender_Demographics_df = pd.DataFrame({"": ["Male", "Female", "Other / Non-Disclosed"],
                                       "Total Count": [len(df_male["SN"].unique()), len(df_female["SN"].unique()), len(df_other["SN"].unique())],
                                       "Percentage of Players": [male_per, female_per, other_per]})

Gender_Demographics_df
```

```python
#Purchasing Analysis (Gender)
#Purchase Count, Average Purchase Price, Total Purchase Value, Avg Total Purchase per Person

f_pc = df_female["Price"].count()
#print(f_pc)
m_pc = df_male["Price"].count()
#print(m_pc)
o_pc = df_other["Price"].count()
#print(o_pc)

f_t = df_female["Price"].sum()
#print(f_t)
m_t = df_male["Price"].sum()
#print(m_t)
o_t = df_other["Price"].sum()
#print(o_t)

f_a = f_t / f_pc
#print(f_a)
m_a = m_t / m_pc
#print(m_a)
o_a = o_t / o_pc
#print(o_a)

f_app = f_t / len(df_female["SN"].unique())
#print(f_app)
m_app = m_t / len(df_male["SN"].unique())
#print(m_app)
o_app = o_t / len(df_other["SN"].unique())
#print(o_app)


Purchasing_Analysis_df = pd.DataFrame({"": ["Female", "Male", "Other / Non-Disclosed"],                                      
                                       "Purchase Count": [f_pc, m_pc, o_pc],
                                      "Average Purchase Price": [f_a, m_a, o_a],
                                      "Total Purchase Value": [f_t, m_t, o_t],
                                      "Avg Total Purchase per Person": [f_app, m_app, o_app]})

Purchasing_Analysis_df

#f_pc.shape

#Remember to round and reformat for dollars

      
```
